# **Exploratory Data Analysis**

## UNIVARIATE ANALYSIS

```python
df.head()
```

```python
#Distribution of Products by Category
sns.countplot(data=df, x='category')
plt.xticks(rotation=45, ha='right')  # Rotate labels for readability
plt.title("Distribution of Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# Top 10 best sellers
df['product'].value_counts().head(10).plot(kind = 'pie', autopct = '%1.1f%%')
plt.show()
```

```python
##Top 10 Brands by Product Count##
sns.countplot(data=df, x='brand', order=df['brand'].value_counts().index[:10])
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Brands by Product Count")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

```python
##Distribution of Sale Prices
sns.histplot(df['sale_price'], kde=True)
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Count")
plt.show()
```

```python
##Boxplot of Sale Prices by Category
sns.boxplot(data=df, x='category', y='sale_price')
plt.xticks(rotation=45, ha='right')
plt.title("Sale Price by Category with Outliers")
plt.xlabel("Category")
plt.ylabel("Sale Price")
plt.show()
```

```python
## Distribution of Product Ratings
sns.histplot(df['rating'], kde=True)
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram of market prices
sns.histplot(df['market_price'], kde=True)
plt.title("Distribution of Market Prices")
plt.xlabel("Market Price")
plt.ylabel("Count")
plt.show()
```

## BIVARIATE ANALYSIS

```python
##1. Scatter Plot for Numerical-Numerical Analysis
# Scatter plot between 'sale_price' and 'rating'
sns.scatterplot(data=df, x='sale_price', y='rating')
plt.title("Sale Price vs. Rating")
plt.xlabel("Sale Price")
plt.ylabel("Rating")
plt.show()

```

```python
## Boxplot for Numerical-Categorical Analysis
# Boxplot for 'sale_price' by 'category'
sns.boxplot(data=df, x='category', y='sale_price')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.title("Sale Price by Category")
plt.xlabel("Category")
plt.ylabel("Sale Price")
plt.show()

```

```python
##3. Heatmap for Categorical-Categorical Analysis
# Create a contingency table for 'category' and 'sub_category'
contingency_table = pd.crosstab(df['category'], df['sub_category'])

# Plot the heatmap
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='coolwarm')  # 'annot' shows numbers, 'fmt' specifies the format
plt.title("Category vs. Subcategory")
plt.xlabel("Subcategory")
plt.ylabel("Category")
plt.show()

```

```python
##Violin Plot for Numerical-Categorical Analysis
# Violin plot for 'rating' by 'category'
sns.violinplot(data=df, x='category', y='rating')
plt.xticks(rotation=45, ha='right')
plt.title("Product Ratings by Category")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.show()

```
